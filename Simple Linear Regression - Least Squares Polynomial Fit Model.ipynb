```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Simple Linear Regression is limited to 1 feature
```

```python
# Corrected file path using raw string
file_path = r"path/to/Advertising.csv"

# Read the CSV file
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
print(df.head())
```

```python
df.head()
```

```python
#Adding a column to calculate the total spent on advertising
df['Total_Spend'] = df['TV'] + df['radio'] + df["newspaper"]
```

```python
df.head()
```

```python
sns.scatterplot(data=df, x='Total_Spend', y='sales')

#Seems to have a positive correlation of spending. The more that's spent the more the sales
```

```python
sns.regplot(data=df, x='Total_Spend', y='sales')
```

```python
X = df['Total_Spend']
y = df['sales']
```

```python
#help(np.polyfit)
#KEY TAKE AWAY: --> Least squares polynomial fit

# y = mx + b : Line for Best Fit
# y = B1x + B0
```

```python
np.polyfit(X, y, deg=1)
#Result is B1x and B0 respectively
```

```python
potential_spend = np.linspace(0, 500, 100)
```

```python
potential_spend
```

```python
predicted_sales = 0.04868788*potential_spend +  4.24302822
```

```python
predicted_sales
```

```python
sns.scatterplot(data=df, x='Total_Spend', y='sales')
plt.plot(potential_spend, predicted_sales, color='red')


#This is the same as regplot
#Can use this to get the expected money needed to be spent to get a certain sales output
```
