```python
#Plot existing data

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16,6))

axes[0].plot(df['TV'], df['sales'], 'o')
axes[0].set_ylabel("Sales")
axes[0].set_title("TV Spend")

axes[1].plot(df['radio'], df['sales'], 'o')
axes[1].set_ylabel("Sales")
axes[1].set_title("Radio Spend")

axes[2].plot(df['newspaper'], df['sales'], 'o')
axes[2].set_ylabel("Sales")
axes[2].set_title("Newspaper Spend")
plt.tight_layout()
```

# Residual Plots

```python
test_residuals = y_test - test_predictions
```

```python
#test_residuals
```

```python
sns.scatterplot(x=y_test, y=test_residuals)
plt.axhline(y=0, color='red', ls='--')

#There is no clear line or curve. This indicates there's something funky with the dataset, therefore best not to use Linear Regression.
```

```python
sns.displot(test_residuals, bins=25, kde=True)
```

```python
#Create a figure and axis to plot  on
fig, ax = plt.subplots(figsize=(6,8), dpi=200)

#Probplot returns the raw  values if needed
#We just want to see the plot, so we assigned these values to _
_ = sp.stats.probplot(test_residuals, plot=ax)
```

```python
final_model = LinearRegression()
```

```python
final_model.fit(X,y)#Fitting to the whole data
```

```python
final_model.coef_
#Give the coefficient for the X in the order of the X Columns
```

```python
X.head()

#Newspaper = -0.00103749 is essentially 0, as the model thinks no matter how much you spend, it's irrelevent to having an effect as there is no impact of the newspapers numbers.
#As it is a negetive number the expectation is that it will negeatively impact your sales

#Based on the numbers it seems the model thinks the Radio spent would have a higher return compared to the rest of the media platforms (TV and Newspaper)
```

```python
#Take the finalmodel and predict over the entire dataset (X)
y_hat = final_model.predict(X)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16,6))

axes[0].plot(df['TV'],df['sales'], 'o')
axes[0].plot(df['TV'],y_hat,'o', color='red')
axes[0].set_ylabel('sales')
axes[0].set_title("TV Spend")

axes[1].plot(df['radio'],df['sales'], 'o')
axes[1].plot(df['radio'], y_hat, 'o', color='red')
axes[1].set_title("Radio Spend")
axes[1].set_ylabel("Sales")

axes[2].plot(df['newspaper'],df['sales'], 'o')
axes[2].plot(df['newspaper'], y_hat, 'o', color='red')
axes[2].set_title("Newspaper Spend")
axes[2].set_ylabel("Newspaper")
plt.tight_layout()

#y_hat is the prediction points
```
